```python
# Load Splice Test Here
data_path = os.path.join(os.getcwd(), "datasets", "splice_test.csv")
splice_test = pd.read_csv(data_path, delimiter=",")

# Your Code goes here:
print("Shape: ", splice_test.shape)
splice_test.head(10)
```

### ========== Question 2.2 ========== 
Convert the categorical attributes into numeric ones by using the [`get_dummies(...)`](https://pandas.pydata.org/pandas-docs/version/1.3.1/reference/api/pandas.get_dummies.html) function from pandas. Make sure to take care of the values `D`, `N`, `S`, `R` (see the [documentation](https://archive.ics.uci.edu/ml/datasets/Molecular+Biology+%28Splice-junction+Gene+Sequences%29) for the data). *Hint: checkout the pandas [`CategoricalDtype`](https://pandas.pydata.org/pandas-docs/version/1.3.1/reference/api/pandas.CategoricalDtype.html)*. Also, make sure to not transform the target variable (`class`).

```python
# Your Code goes here:
pd.set_option('display.max_columns', None)

#Convert N class labels to prevent them getting changed
splice_train.loc[splice_train["class"] == "N", "class"] = "P"
splice_test.loc[splice_test["class"] == "N", "class"] = "P"

#Replace ambigious terms with NaN
splice_train = splice_train.replace(['D', 'N', 'S', 'R'] ,np.NaN)
splice_test = splice_test.replace(['D', 'N', 'S', 'R'] ,np.NaN)

#Remove all rows containing ambigious terms
splice_train = splice_train.dropna()
splice_test = splice_test.dropna()

#Return N values to their original state in the class column
splice_train.loc[splice_train["class"] == "P", "class"] = "N"
splice_test.loc[splice_test["class"] == "P", "class"] = "N"

#Take dummies, excluding the class column
Dummied_train_values = pd.get_dummies(splice_train.drop(['class'], axis = 1))
Dummied_test_values = pd.get_dummies(splice_test.drop(['class'], axis = 1))

#Stick the class column back on
X = pd.DataFrame(Dummied_train_values)
y = pd.DataFrame(Dummied_test_values)

X['class'] = splice_train['class']
y['class'] = splice_test['class']
```

### ========== Question 2.9 ==========
Now we want to explore the effect of the `k` parameter. To do this, train the classifier multiple times, each time setting the KNN option to a different value. Try `5`, `10`, `50`, `100`, `200`, `500`, `1000`, `1500` and `2000` and test the classifier on the test set. How does the k parameter effect the results? *Hint: Consider how well the classifier is generalising to previously unseen data, and how it compares to the dumb prediction accuracy.*

```python
# Your Code goes here:
results = pd.DataFrame(columns = ["neighbours", "accuracy", "log_losses"])

#Arrays for storage
accuracies = []
num_neighbours = [5, 10, 50, 100, 200, 500, 1000, 1500, 2000]
log_losses = []

#Small function to run classifier for each different value of k
def run_nearest_neighbour(num_n):
    clf_K = KNeighborsClassifier(n_neighbors=num_n)
    clf_K = clf_K.fit(X_train, y_train)
    accuracy = clf_K.score(X_test, y_test) * 100
    predictions = clf_K.predict_proba(X_test)
    loss = log_loss(y_test, predictions)
    print("Testing Accuracy with {} neighbours:  {:.2f}".format(num_n, accuracy))
    accuracies.append(accuracy)
    log_losses.append(loss)
    return clf_K
    
for i in num_neighbours:
    run_nearest_neighbour(i)

#concatenate into a dataframe
results["neighbours"] =  num_neighbours
results['accuracy'] = accuracies
results['log_losses'] = log_losses

results.head(9)
```

***Your answer goes here:***

The K parameter essentially increases the level of complexity in the solution. Up to 1000 neighbours, this generalises to testing accuracy well, before becoming absurdly large and becoming detrimental to the point 2000 neighbours performs almost worse than the dummy.

```python
# Your Code goes here:
#Plot with matplotlib
plt.plot(results['neighbours'], results['accuracy'])
plt.xlabel("Number of Neighbours")
plt.ylabel("Accuracy %")
```

***Your answer goes here:***

From the plot we can see the optimum tested number of neighbours is 1000, before sharply declining in accuracy.

### ========== Question 2.11 ==========
Select best value for `k` from Questions 2.9 and 2.10 and plot the normalised confusion matrix on the test set (you may use the provided function). Then plot the confusion matrix for a 5-nearest neighbour classifier. What do you observe?

```python
def plot_confusion_matrix(cm, classes=None, title='Confusion matrix'):
    """Plots a confusion matrix."""
    if classes is not None:
        sns.heatmap(cm, xticklabels=classes, yticklabels=classes, vmin=0., vmax=1., annot=True)
    else:
        sns.heatmap(cm, vmin=0., vmax=1.)
    plt.title(title)
    plt.ylabel('True label')
```

```python
# Your Code goes here:
#Plot for the 1000 neighbour classifier
clf_K = run_nearest_neighbour(1000)
confusion = confusion_matrix(y_test, clf_K.predict(X_test))
normalized_matrix = confusion/ confusion.astype(np.float).sum(axis=1)
plot_confusion_matrix(normalized_matrix, ['True', 'False'])
```

```python
#Plot with a 5 neighbour classifier
clf_K = run_nearest_neighbour(5)
confusion = confusion_matrix(y_test, clf_K.predict(X_test))
normalized_matrix = confusion/ confusion.astype(np.float).sum(axis=1)
plot_confusion_matrix(normalized_matrix, ['True', 'False'])
```

***Your answer goes here:***

With less neighbours, the 5-neighbour classifier is far more likely to just guess the most frequent class of N.

### ========== Question 2.12 ==========
Read about the [logarithimic loss](http://scikit-learn.org/0.24/modules/generated/sklearn.metrics.log_loss.html) (or cross-entropy loss). It is often the error metric used when we are trying to optimise classification models.

This metric takes as input the true labels and the estimated probability distributions (bernouli or multinomial). It makes sense to use this metric when we are interested not only in the predicted labels, but also in the confidence with which these labels are predicted.

For instance, think of the situation where you have a single test point and two classifiers. Both classifiers predict the label correctly, however classifier A predicts that the test point belongs to the class with probability 0.55, whereas classifier B predicts the correct class with probability 0.99. Classification accuracy would be the same for the two classifiers (why?) but the `log_loss` metric would indicate that classifier B should be favoured.

Produce a scatter plot similar to the one in Question 2.10 but this time show `log_loss` on your y axis. Which value for `k` would you pick if `log_loss` was the error metric? Comment on why this might happen, and which metric would be a better evaluator of performance.

***Your answer goes here:***

Classification accuracy would be the same for both, because the accuracy metric doesn't take confidence into consideration for its calculation. if log loss was the only metric to decide on, then 5 neighbours would be picked as it was the most confident. This may happen because despite being less accurate, the ones it got correct (when usually picking the majority class) it gets with a high confidence, bringing its log loss down. in the case like this where the classes are highly imbalanced, f1 score would be an ideal indicator of performance, but accuracy would be better than log loss.
